```
# CELL 2: Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_curve, auc, precision_score, recall_score, f1_score
)
from sklearn.utils.class_weight import compute_class_weight

from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

np.random.seed(42)
tf.random.set_seed(42)
```

```
# CELL 3: Load real dataset

url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)

print(df.shape)
df.head()
print(df.head().T)
```

```
# CELL 4: Clean data

df = df.drop("customerID", axis=1)

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())

df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})

print(df["Churn"].value_counts())
print(df["Churn"].value_counts(normalize=True) * 100)
```

```
# CELL 5: Split features and target

X = df.drop("Churn", axis=1)
y = df["Churn"]

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```
# CELL 6: Preprocessing

try:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_cols),
        ("cat", encoder, categorical_cols)
    ]
)

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print(X_train_processed.shape)
print(X_test_processed.shape)
```

```
# CELL 7: ANN model function

def build_ann(input_dim, layers=[64, 32], dropout_rate=0.3, learning_rate=0.001):
    model = Sequential()

    model.add(Input(shape=(input_dim,)))

    for neurons in layers:
        model.add(Dense(neurons, activation="relu"))
        model.add(BatchNormalization())
        model.add(Dropout(dropout_rate))

    model.add(Dense(1, activation="sigmoid"))

    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model
```

```
# CELL 8: Baseline model

input_dim = X_train_processed.shape[1]

baseline_model = build_ann(
    input_dim=input_dim,
    layers=[64, 32],
    dropout_rate=0.3,
    learning_rate=0.001
)

early_stop_baseline = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

baseline_history = baseline_model.fit(
    X_train_processed,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop_baseline],
    verbose=1
)

baseline_probs = baseline_model.predict(X_test_processed).ravel()
baseline_preds = (baseline_probs >= 0.5).astype(int)

baseline_accuracy = accuracy_score(y_test, baseline_preds)

print("Baseline Accuracy:", baseline_accuracy)
print(classification_report(y_test, baseline_preds, target_names=["No Churn", "Churn"], zero_division=0))
```

```
# CELL 9: Class weighted model

classes = np.unique(y_train)

weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))
print(class_weights)

weighted_model = build_ann(
    input_dim=input_dim,
    layers=[128, 64, 32],
    dropout_rate=0.25,
    learning_rate=0.0005
)

early_stop_weighted = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

weighted_history = weighted_model.fit(
    X_train_processed,
    y_train,
    validation_split=0.2,
    epochs=150,
    batch_size=32,
    callbacks=[early_stop_weighted],
    class_weight=class_weights,
    verbose=1
)

weighted_probs = weighted_model.predict(X_test_processed).ravel()
weighted_preds = (weighted_probs >= 0.5).astype(int)

weighted_accuracy = accuracy_score(y_test, weighted_preds)

print("Weighted Accuracy:", weighted_accuracy)
print(classification_report(y_test, weighted_preds, target_names=["No Churn", "Churn"], zero_division=0))
```

```
# CELL 10: SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train_processed, y_train)

print("Before SMOTE:")
print(y_train.value_counts())

print("\nAfter SMOTE:")
print(pd.Series(y_train_smote).value_counts())
```

```
# CELL 11: Tune ANN models

configs = [
    {"layers": [64, 32], "dropout": 0.20, "lr": 0.001, "batch": 32},
    {"layers": [128, 64], "dropout": 0.25, "lr": 0.001, "batch": 32},
    {"layers": [128, 64, 32], "dropout": 0.30, "lr": 0.0005, "batch": 32},
    {"layers": [256, 128, 64], "dropout": 0.35, "lr": 0.0005, "batch": 64},
]

results = []

best_model = None
best_history = None
best_accuracy = 0
best_config = None
best_probs = None
best_preds = None

for config in configs:
    print("Training:", config)

    model = build_ann(
        input_dim=input_dim,
        layers=config["layers"],
        dropout_rate=config["dropout"],
        learning_rate=config["lr"]
    )

    early_stop = EarlyStopping(
        monitor="val_loss",
        patience=10,
        restore_best_weights=True
    )

    history = model.fit(
        X_train_smote,
        y_train_smote,
        validation_split=0.2,
        epochs=150,
        batch_size=config["batch"],
        callbacks=[early_stop],
        verbose=0
    )

    probs = model.predict(X_test_processed).ravel()
    preds = (probs >= 0.5).astype(int)

    acc = accuracy_score(y_test, preds)
    precision = precision_score(y_test, preds, zero_division=0)
    recall = recall_score(y_test, preds, zero_division=0)
    f1 = f1_score(y_test, preds, zero_division=0)

    results.append({
        "layers": str(config["layers"]),
        "dropout": config["dropout"],
        "learning_rate": config["lr"],
        "batch_size": config["batch"],
        "accuracy": acc,
        "precision_churn": precision,
        "recall_churn": recall,
        "f1_churn": f1
    })

    if acc > best_accuracy:
        best_accuracy = acc
        best_model = model
        best_history = history
        best_config = config
        best_probs = probs
        best_preds = preds

results_df = pd.DataFrame(results)
results_df
```

```
# CELL 12: Final results

print("Baseline Accuracy:", baseline_accuracy)
print("Weighted Accuracy:", weighted_accuracy)
print("Best Tuned Accuracy:", best_accuracy)
print("Best Config:", best_config)

print("\nClassification Report:")
print(classification_report(y_test, best_preds, target_names=["No Churn", "Churn"], zero_division=0))
```

```
# CELL 13: Confusion matrix

cm = confusion_matrix(y_test, best_preds)

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["No Churn", "Churn"],
    yticklabels=["No Churn", "Churn"]
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Tuned ANN")
plt.show()
```

```
# CELL 14: ROC curve

fpr, tpr, thresholds = roc_curve(y_test, best_probs)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Tuned ANN")
plt.legend()
plt.grid(True)
plt.show()

print("ROC AUC:", roc_auc)
```

```
# CELL 15: Accuracy curve

plt.figure(figsize=(7, 5))
plt.plot(best_history.history["accuracy"], label="Training Accuracy")
plt.plot(best_history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```
# CELL 16: Loss curve

plt.figure(figsize=(7, 5))
plt.plot(best_history.history["loss"], label="Training Loss")
plt.plot(best_history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```
# CELL 17: Final report summary

final_accuracy = accuracy_score(y_test, best_preds)
final_precision = precision_score(y_test, best_preds, zero_division=0)
final_recall = recall_score(y_test, best_preds, zero_division=0)
final_f1 = f1_score(y_test, best_preds, zero_division=0)

print("FINAL SUMMARY")
print("-------------")
print(f"Baseline ANN Accuracy: {baseline_accuracy:.4f}")
print(f"Class Weighted ANN Accuracy: {weighted_accuracy:.4f}")
print(f"Best Tuned ANN Accuracy: {final_accuracy:.4f}")
print(f"Churn Precision: {final_precision:.4f}")
print(f"Churn Recall: {final_recall:.4f}")
print(f"Churn F1-score: {final_f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")

print("""
Conclusion:
The IBM Telco Churn dataset is imbalanced, with about 73% No-Churn and 27% Churn.
Therefore, 0.95 accuracy is not realistically expected without overfitting or data leakage.
A result around 0.78 to 0.83 accuracy is more realistic and defensible.
For churn prediction, recall and F1-score for the Churn class are more meaningful than accuracy alone.
""")
```
